# Life Expectancy and GDP

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

life_expectancy = pd.read_csv("all_data.csv") #reading the file
life_expectancy.head()
```

Lets first take a look at life expectancy in the six nations over time

```python
fig1 = plt.subplots(figsize=(7, 6))
sns.lineplot(data = life_expectancy,x = 'Year', y = 'Life expectancy at birth (years)', hue = 'Country')
plt.savefig("overtime.png")
plt.show()
plt.clf()
```

We can see that life expectancy sagnificantly risen over 15 year in Zimbabwe and had a steady slower growth in otther countries. We can take a closer look for each contry:

```python
plt.figure()
ax1 = plt.subplot(3,2,1)
dataChile = life_expectancy[life_expectancy.Country == 'Chile']
ax1.set_ylabel('LEBY')
ax1.set_title('Chile')
sns.lineplot(data = dataChile, x = 'Year', y = 'Life expectancy at birth (years)')

ax2 = plt.subplot(3,2,2)
dataChina = life_expectancy[life_expectancy.Country == 'China']
ax2.set_ylabel('LEBY')
ax2.set_title('China')
sns.lineplot(data = dataChina, x = 'Year', y = 'Life expectancy at birth (years)')

ax3 = plt.subplot(3,2,3)
dataGermany = life_expectancy[life_expectancy.Country == 'Germany']
ax3.set_ylabel('LEBY')
ax3.set_title('Germany')
sns.lineplot(data = dataGermany, x = 'Year', y = 'Life expectancy at birth (years)')

ax4 = plt.subplot(3,2,4)
dataMexico = life_expectancy[life_expectancy.Country == 'Mexico']
ax4.set_ylabel('LEBY')
ax4.set_title('Mexico')
sns.lineplot(data = dataMexico, x = 'Year', y = 'Life expectancy at birth (years)')

ax5 = plt.subplot(3,2,5)
dataUSA = life_expectancy[life_expectancy.Country == 'United States of America']
ax5.set_ylabel('LEBY')
ax5.set_title('USA')
sns.lineplot(data = dataUSA, x = 'Year', y = 'Life expectancy at birth (years)')

ax6 = plt.subplot(3,2,6)
dataZimbabwe = life_expectancy[life_expectancy.Country == 'Zimbabwe']
ax6.set_ylabel('LEBY')
ax6.set_title('Zimbabwe')
sns.lineplot(data = dataZimbabwe, x = 'Year', y = 'Life expectancy at birth (years)')

plt.tight_layout()
plt.savefig("lebybycountry.png")
plt.show()
plt.clf()
```

Next we would want to understand if GDP has increased over time in the six nations

```python
sns.lineplot(data = life_expectancy,x = 'Year', y = 'GDP', hue ='Country')
plt.show()
plt.clf()
```

We can notice sagnificant GDP rise in United states and China.

```python
sns.scatterplot(data = life_expectancy,x = 'Life expectancy at birth (years)', y = 'GDP', hue ='Country')
plt.show()
plt.clf()
```

While there was sagnificant GDP growth in China and United states, Life expectancy growth was changing as slowly as in countries who had slow rise of GDP. We can also notice that China had always 5 years life expectancy increase between 2000 and 2006, while GDP that time was still growing slow, and strating 2006 with rapid rise of GDP, life expectancy in China was growing very slowly.
